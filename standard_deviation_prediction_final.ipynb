```python
data = pd.read_csv('CC2020_train_final.csv')
```

```python
data.head(3)
```

```python
X = data.drop(['YEAR', 'LOCATION'], axis=1)
```

```python
len(X)
```

```python
clean_data = X.groupby(['INBRED', 'INBRED_CLUSTER','TESTER', 'TESTER_CLUSTER' ],as_index=False).agg(np.std, ddof=1)
```

```python
clean_data.head(5)
```

```python
len(clean_data)
```

```python
y = clean_data.YIELD
```

```python
data = clean_data.drop(['YIELD'], axis=1)
```

```python
data.head(5)
```

```python
X = data
```

```python
label_enc = preprocessing.LabelEncoder()
```

```python
X['INBRED'] = label_enc.fit_transform(X['INBRED'])
X['INBRED_CLUSTER'] = label_enc.fit_transform(X['INBRED_CLUSTER'])
X['TESTER'] = label_enc.fit_transform(X['TESTER'])
X['TESTER_CLUSTER'] = label_enc.fit_transform(X['TESTER_CLUSTER'])
```

```python
X.head(3)
```

```python
#notice how these are all same.. number varies within same location or year
```

```python
enc = preprocessing.OneHotEncoder()
```

```python
X = enc.fit_transform(X)
```

#Classifiers

```python
cat = CatBoostRegressor(iterations=2000, learning_rate=0.05, depth=16, l2_leaf_reg= 2)
```

```python
cat.fit(X, y)
```

```python
test = pd.read_csv('CC2020_test_final.csv')
```

```python
X_test = test.drop(['YIELD'], axis=1)
```

```python
X_test['INBRED'] = label_enc.fit_transform(X_test['INBRED'])
X_test['INBRED_CLUSTER'] = label_enc.fit_transform(X_test['INBRED_CLUSTER'])
X_test['TESTER'] = label_enc.fit_transform(X_test['TESTER'])
X_test['TESTER_CLUSTER'] = label_enc.fit_transform(X_test['TESTER_CLUSTER'])
```
